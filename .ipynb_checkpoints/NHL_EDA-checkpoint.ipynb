# Carolina Hurricanes through the years

```python
#Set up
import requests as req
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def jprint(obj):
    # create a formatted string of the Python JSON object
    text = json.dumps(obj, sort_keys=True, indent=4)
    print(text)
```

```python
response = req.get("https://statsapi.web.nhl.com/api/v1/teams/12")
jprint(response.json())
```

First year of canes: 1997-1998

```python
import datetime
#get all the seasons that the canes have played, starting with 1997
year1 = 1997
year2 = 1998
seasons = []
while year2 <= datetime.datetime.now().year:
    season = str(year1) + str(year2)
    seasons.append(season)
    year1 = year2
    year2 += 1
print(seasons)
```

```python
#use the seasons to make a list of all the gameID's (gamePk)
#https://statsapi.web.nhl.com/api/v1/schedule?teamId=12&season=19971998 <-- returns canes 98-99 schedule
for szn in seasons:
    test = req.get("https://statsapi.web.nhl.com/api/v1/schedule?teamId=12&season=" + szn)
    print("--------------" + str(szn))
    for i in range(len(test.json()["dates"])):
        jprint(test.json()["dates"][i]["games"][0]["gamePk"])
    #break #for now
```
